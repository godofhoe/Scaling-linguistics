## Compose N-gram corpus, B.txt, by shuffling words in A.txt into characters and randomly group them into new N-gram words

```python
import random 
import bisect 
import math 
from functools import reduce
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from count import*
%matplotlib inline
```

```python
def read_script(inp_file_name, sp = " "):
    """Read a fake script and return a list contains all the characters in the script.
    """
    text =  open(inp_file_name,"r",encoding = 'utf-8')
    string_list = []

    for line in text:
        l = line.split()
        for word in l:
            for char in list(word):
                string_list.append(char)

    text.close()        
    return string_list
```

```python
def generate_DDWang_text(inp_file_name, out_file_name, word_len):
    """reads inp_file, generateing word_len-gram script and saving it as out_file.
    ex: generate_DDWang_text(A.txt, B.txt, 3) then B.txt is composed of 3-gram words

    Parameters
    ---
    file_name: str
        the name of a file containing novel texts
    
    word_len: int
        an interger >= 0
        
    Returns
    ---
    None
    
    Raises
    ---
    AssertionError

    """
    assert word_len >= 0
    f2 = open(out_file_name , 'w', encoding = 'utf-8')
    string_list = read_script(inp_file_name, sp = " ")
    random.shuffle(string_list)
    list_2 = []
    tmp = ''
    for e in string_list:
        tmp = tmp + e
        if len(tmp) == word_len:
            list_2.append(tmp)
            tmp = ''
   
    word_count = 0
    for item in list_2:
        if word_count < 20:
            f2.write(item+" ")
            word_count += 1
        else:
            word_count = 0
            f2.write(item+"\n")
    f2.close()
    print("A DDWang fake script is successfully created!")
    return None
```

```python
filename = "ddwang_2gram.txt" #user define
inp_file = 'hypocritical holy war.txt'
#Generate word_len-gram corpus
generate_DDWang_text(inp_file, out_file_name = filename, word_len = 2)
```
